```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-hf",)
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-hf", low_cpu_mem_usage = True,
                                             load_in_8bit = True)
```

```python
inputs = tokenizer("Hello write 3 sentences on Malabari Biriyani and Mohanlal",
                            return_tensors = "pt")

outs = model.generate(**inputs.to('cuda'))
```

```python
tokenizer.decode(outs)
```
